```python
# COMPLETE THESE TASKS
# 1. look into the documentation for `np.matmul`, `np.dot`, `@`, and `*` and explain their difference(s)
# 2. create an example/examples to explain their differences

A = np.random.rand(2,2)
B = np.random.rand(2,2)
a = np.random.rand(2)
b = np.random.rand(2)

print("Test matrices:")
print(f'A = {A}')
print(f'B = {B}')
print(f'a = {a}')
print(f'b = {b}')

print("Testing elementwise multiplication:")
# "A*B": Elementwise multiplication, A and B must have compatible shapes and if they don't, broadcasting rules apply
print(f'A*B = {A*B}')
print(f'a*b = {a*b}')
print(f'A*b = {A*b}')

print("Testing dot multiplication:")
# "np.dot(A, B)": returns dot product for 1d matrices, matrix product for Nd matrices, and matrix vector product for Nd 1d product
print(f'np.dot(A,B) = {np.dot(A,B)}')
print(f'np.dot(a,b) = {np.dot(a,b)}')
print(f'np.dot(A,b) = {np.dot(A,b)}')

print("Testing matmul:")
# "np.matmul(A, B)": simular to np.dot but 1d, 1d product leads to error and supports batch multiplication
print(f'np.matmul(A,B) = {np.matmul(A,B)}')
#print(f'np.matmul(a,b) = {np.matmul(a,b)}') # causes error
print(f'np.matmul(A,b) = {np.matmul(A,b)}')

print("Testing @:")
# "@": shorthand for matmul, cleaner and doesn't cause error for 1d, 1d dot product
print(f'A@B = {A@B}')
print(f'a@b = {a@b}')
print(f'A@b = {A@b}')
```

# Eigenvalues and Eigenvectors

```python
K = np.array([[ 1,-1, 0]
             ,[-1, 2,-1]
             ,[ 0,-1, 1] ])
vals, vecs = np.linalg.eigh(K)
vals, vecs
```

```python
# COMPLETE THIS TASK
# In the above cell, which eigenvector is associated with which eigenvalue?
print((K@vecs[:, 0]) - vals[0]*vecs[:, 0]) # = [0, 0, 0], lambda_1 = 1e-16
print(K@vecs[:, 1] - vals[1]*vecs[:, 1]) # = [0, 0, 0], lambda_2 = 1
print(K@vecs[:, 2] - vals[2]*vecs[:, 2]) # = [0, 0, 0], lambda_3 = 3
```

Quadratic Form: Diatomic Molecule Model
We can model a diatomic molecule (O2
 or N2
, for example) as two masses connected by a spring. Motion is restricted to the bond axis.

The (quadratic) potential energy is: V = 0.5 * x.T @ K @ x  where x = (x1, x2), the coordinates of the two atoms.

Constructing a stiffness/force-constant matrix

```python
k = 2450 # the force constant in an appropriate [energy]/[length] units
K = np.array([[ k,-k]
             ,[-k, k]])
K
```

Diagonalize to find frequencies and normal mode displacement patterns

```python
vals, vecs = np.linalg.eigh(K)
vals, vecs
```


Vibrational frequencies (ignoring mass)

```python
omega = np.sqrt(vals)
omega
```

Unitary transformations with eigenvector matrices
If the matrix is not singular, we can find eigenvalues and eigenvectors. The matrix of eigenvectors U
 forms a unitary transformation.

```python
# COMPLETE THESE TASKS
# Show that U@U.T and U.T@U are 2x2 identity matrices
# Note that normally we would need to use the `.conj().T` attribute to find the conjugate transpose matric, but our matrix is real.

#L = np.random.rand(2,2)
eigens, U = np.linalg.eigh(K)
print(U@U.T)
print(U.T@U)
```

```python
# COMPLETE THIS TASK
# From the eigenvalue equation M v = λ v, it follows that for U = [v1, v2], then M U = λ U.
# Because λ is diagonal and U.T@U = np.eye(2), it follows that λ = U.T @ K @ U.
# Confirm this behavior.

LAMBDA = np.diag(eigens)
print(U.T@K@U)
print(LAMBDA)

```

```python
# COMPLETE THIS TASK
# Use the eigenvectors to eigenvalues to interpret the two modes in this model of a diatomic molecule

vec1 = U[:, 0]
vec2 = U[:, 1]

print(vec1)
print(K@vec1) # goes to 0
# the first eigenvalue indicates that the frequency is 0 first this mode

print(vec2)
print(K@vec2)
# The large output of the eigenvector multiplication indicates the second mode frequency is very high
# The frequency of 70 in omega is consistent with this.
print(omega)
```
